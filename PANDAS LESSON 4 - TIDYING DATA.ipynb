```python

#this is a single column stored in multiple columns 

#ebola data set with the same problem
ebola = pd.read_csv('ebola_country_timeseries.csv')
ebola.head()

#cases and deaths of various countries 
#first get to know whether it is case or death and then the country 

ebola_melt = pd.melt(ebola, id_vars = ['Date','Day'], var_name = 'cd_country', value_name = 'count')
ebola_melt

#if we want take a column and treat it as a vector of strings we use .str and then we can do standard string ops
ebola_melt['cd_country'].str
#now use string.split to split cases and deaths 
var_split = ebola_melt['cd_country'].str.split('_')

#var_split has a list of lists
#if we want the 1st element of all lists
#if we want each element in the list which is a part of a series 
#str makes sure it acts in a vectorized manner 
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)


#assigning new columns in our data frame
ebola_melt['Status'] = status_values
ebola_melt['Country'] = country_values
```

```python
type(ebola_melt['cd_country'].str.split('_')[0])

#each element is a list and a bunch of list is a pandas series 
#first element of the list is status and the second is country 

variable_split = ebola_melt['cd_country'].str.split('_', expand = True)
variable_split
#instead each element as a list of series, it returns a data frame

#renaming columns
variable_split.columns = ['Status1','Country1']
variable_split

#now concatenate
ebola_clean = pd.concat([ebola_melt,variable_split],axis = 1)
ebola_clean.head()
```

```python
# another cleaning problem 
#varibales both in row and column 
weather = pd.read_csv('weather.csv')
weather.head()
#how to know there is a problem - when 2 or more rows are almost identical minus a few minor details 
#2 3 duplicated except element series. so a lot of duplicates in this data 
```

```python


#first step to fix - melt 
weather_melt = pd.melt(weather, id_vars = ['id','year','month','element'], var_name = 'Day', value_name = 'temp')
weather_melt.head()
#we can also do weather.melt
#now you can see duplicates clearly - because of element 
```

```python


#next step - opposite of melt - pivot into columns 
#take each unique value of element and turn that into a seperate column 
#for that pivot table - index - columns that you want to keep the same , columns to pivot up that column , values will
#tell what values you want to populate new formed columns with
weather_tidy = weather_melt.pivot_table(index = ['id','year','month','Day'], columns = 'element', values = 'temp')
```

```python
weather_tidy = weather_tidy.reset_index()
weather_tidy.head(20)

# weather_tidy['Day'].str.get()

#in pivot table if duplicates are generated, you can replace the value with aggfunc with mean min max etc 

#now we have to convert days to integer 
```
