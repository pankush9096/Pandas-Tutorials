```python
#merging data sets (table joints )

#loading data 
person = pd.read_csv('survey_person.csv') #people on this arctic exploration 
site = pd.read_csv('survey_site.csv') #site where they have taken readings 
survey = pd.read_csv('survey_survey.csv') #where they performed a set of readings at a particular site
visited = pd.read_csv('survey_visited.csv') #who visited what site on what day 

visited_subset = visited.iloc[[0,2,6],:]
print(visited_subset)

print(visited)

#well put all these tables together 
#kinds of merges - 1:1 when we have all unique values and want to merge 
#                - if one of them have duplicated values --> many - 1 or 1 - many join  
#the dataset which doesnt have duplicated ends up getting duplicated in the one that does have duplicates 
#                - many many --> both have duplicates - cartisian product 
```

```python
#all these merges happen in the same way. Just need to be aware what is in the dataset 
#in jupyter, shift tab gives you the function signature
#inner join keeps only the value that matches - left join - keep all the values on the left side 
#inner by default 
#on parameter if the columns you want to do the match on have the same name 

# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,left_index=False, right_index=False, sort=True)

# left − A DataFrame object.

# right − Another DataFrame object.

# on − Columns (names) to join on. Must be found in both the left and right DataFrame objects.

# left_on − Columns from the left DataFrame to use as keys.
# Can either be column names or arrays with length equal to the length of the DataFrame.

# right_on − Columns from the right DataFrame to use as keys.
# Can either be column names or arrays with length equal to the length of the DataFrame.

# left_index − If True, use the index (row labels) from the left DataFrame as its join key(s).
# In case of a DataFrame with a MultiIndex (hierarchical), the number of levels must match the number of join keys from the right DataFrame.

# right_index − Same usage as left_index for the right DataFrame.

# how − One of 'left', 'right', 'outer', 'inner'.
# Defaults to inner. Each method has been described below.

# sort − Sort the result DataFrame by the join keys in lexicographical order.
# Defaults to True, setting to False will improve the performance substantially in many cases.


o2o = pd.merge(site,visited_subset,left_on='name',right_on ='site')
o2o

#left_on specifying what column in the left data frame i am using as the key - we can also give multiple keys 
#right_on specifying what column in the right data frame i am using as the key

#another way of merging
```

```python
m = site.merge(visited,left_on = 'name', right_on = 'site', sort='FALSE')
m
```

```python
#we can make a data frame using a python dictionary 
#key becomes the column  , values become the entire row of data 

df1 = pd.DataFrame({'a': [1,1,1,2,2],
                   'b': [10,20,30,40,50]})

df2 = pd.DataFrame({'a1': [1,1,2,2,3],
                   'b1' : [100,200,300,400,500]})
df1
```

```python
df2
```

```python
#duplicate values in both sets 
#cartisian product - many - many 
df1.merge(df2,left_on = 'a', right_on = 'a1', how='left')

#got back a lot more rows than expected 
#creates every pair wise combination 
#to not do cartisian, remove duplicates 

#to check duplicates in your data 
# df1.duplicated('a')
```

```python
import pandas as pd
df = pd.DataFrame({"A":["foo", "foo", "foo", "bar"], "B":[0,1,1,1], "C":["A","A","B","A"]})
df
```

```python
df.drop_duplicates(subset=['A'], keep='first') # keep first encountered value and drop other
```

```python
df.drop_duplicates(subset=['B'], keep='last') # keep last value and remove other
```

```python
df.drop_duplicates(subset=['B','C'], keep=False) # remove all value having dups
```

```python
df.drop_duplicates(['A','B','C']) # remove the dup pair of 3 col values
```
