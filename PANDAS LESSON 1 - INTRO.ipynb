```python
#3 ways to subset rows

#1st. loc. loc would pull out the row as well the corresponding column
#loc pulls out rows by what it was labelled (index same as label in this case )
df.loc[56]

#[-1] wont work since it works on label and not index 
df.loc[0]

#use iloc to use -1 since iloc works on indexes and not labels
#pull out from index -1 and not row labelled -1 
df.iloc[-1]

#pulling out mulitple rows 
df.loc[[0,10,999]]

#subsetting  both rows and columns. first entry row, second entry column 
df.loc[[0,10,999],['country','pop','year']]

#names in iloc wont work because iloc works on indexes and not labels
df.iloc[[0,10,999],[0,2,4]]

#standard slicing for iloc would work
df.iloc[3:10,0:2]
```

```python
#each column ends up becoming a pandas series. these series all have various built in methods 
lifexp_mean = df['lifeExp'].mean()
lifexp_mean
```

```python
#subsetting data by using loc. loc first entry subset rows, second subset columns 
df.loc[df['lifeExp'] > lifexp_mean,:] 
```

```python
#in pandas we can do group/ aggregate statistics 
#we want to know average life expectancy for each year
#subset data frame by each unique value for year 

df.groupby('year')
#makes a special data frame with a marker saying group it by year and then perform operations 

#returns a groupby object 

df.groupby('year')['lifeExp'].mean()
#partitions the data into each year, applies mean and then stitches it back together - split --> apply --> combine 
#this is the corner stone of how distributed computing works

#we can do the same by pulling out multiple columns 
df.groupby('year')[['lifeExp','gdpPercap']].mean()

#applying multiple groupby is also possible - by putting it in square brackets 
df.groupby(['year', 'continent'])[['lifeExp','gdpPercap']].mean().reset_index()
#in this you'll see a weird indentation which makes indexing difficult so to flatten it we use reset_index()
#not after resetting it becomes our usual data frame without any heirarchy 

#by using \ we can tell python one line is written in multiple lines  or put it in paranthesis. that way python will
#know to treat it under one command 

#nunique lets you count the number of unique values
#df.groupby('continent')['country'].nunique()
```

```python
#setting a variable to groupby 
gyle = df.groupby('year')['lifeExp'].mean()

#we can plot this varibale too
import matplotlib as plt
%matplotlib inline
#whatever plot gets rendered by plt display underneath this cell in jupyternotebook 
#if not we would use plt.show() whichll open a popup window 
  
gyle.plot()
type(gyle) # --> pandas series object - it has labels which are years and its corresponding values --> mean life exp

#if we want this to be a data frame 
gyle.reset_index()
#now we have a column for year and a column for mean life exp 
df_new = gyle.reset_index()
#each data frame has a to function which saves to something
df_new.to_csv('lifeExp_by_year.csv')
#if we dont want the index, put the parameter index = false inside to csv 
#if index is not special , delete it 
```
